```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/sentimentdataset.csv')
```

```python
df.head(3)
```

```python
df.info()
```

```python
df = df.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
```

```python
df[['Date', 'Time']] = df['Timestamp'].str.split(' ', expand=True)
df = df.drop('Timestamp', axis=1)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
```

```python
df['Retweets'] = df['Retweets'].astype(int)
df['Likes'] = df['Likes'].astype(int)
```

```python
df.isnull().sum()
```

```python
# setting float format into only 2 decimals points
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df.describe()
```

```python
df['Hashtags'] = df['Hashtags'].str.replace('#', '')
```

```python
df[['Hashtags0', 'Hashtags1', 'Hashtags2']] = df['Hashtags'].str.split(' ', expand=True, n=2)
# Adding a third column 'Hashtags3' to capture all hashtags beyond the first two.
# Using n=2 to limit the split to two splits and avoid None values in the first two columns.
```

```python
df = df.drop(['Hashtags', 'Hashtags0'], axis=1)
```

```python
df['Hashtags2'] = df['Hashtags2'].str.strip(' ')
df['Hashtags1'] = df['Hashtags1'].str.strip(' ')
df['Sentiment'] = df['Sentiment'].str.strip(' ')
df['Country'] = df['Country'].str.strip(' ')
df['Platform'] = df['Platform'].str.strip(' ')
df['User'] = df['User'].str.strip(' ')
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.reset_index(drop = True)
```

```python
# download cleaned dataset in csv format
df.to_excel('Social Media Sentiments Analysis.xlsx')
```
